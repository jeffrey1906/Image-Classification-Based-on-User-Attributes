```python
# to get the shareable column using clustering technique
```

```python
#pandas to read the dataframe
import pandas as pd
```

```python
#reading excel file
train = pd.read_excel('data5.xls')
```

```python
#display first 5 rows of dataframe
train.head()
```

```python
#importing KMeans library
from sklearn.cluster import KMeans
```

```python
#separating into 2 clusters
k = KMeans(n_clusters=2,n_jobs=-1)
```

```python
k.fit(train)
```

```python
#predicting the clusters
predict = k.predict(train)
```

```python
#displaying the cluster values
predict
```

```python
#accuracy of the clustering
k.inertia_
```

```python
#adding predicted value to the dataset
train['shareable'] = predict
```

```python
#displaying the training data
train
```

```python
#importing ExcelWriter for Dataframe to Excel export
from pandas import ExcelWriter
```

```python
#saved to data6.xlsx
writer = ExcelWriter('data6.xlsx')
train.to_excel(writer,'Sheet1')
writer.save()
```
